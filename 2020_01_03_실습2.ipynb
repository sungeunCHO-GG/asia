# 컬럼 순서가 바뀐 새로운 wine_df2 만들기
-pH_5, sulphates, alchol, quality_6
-chlorides, free_sulfur_dioxide, total_sulfur_dioxide, density
-fixed_acidity, volatile_acidity, citic_acid, residual_sugar

```python
# 컬럼명 바꾸기(2개 컬럼만)

df = df.rename({'pH':'ph_5','quality':'quality_6'},axis='columns')
df.head(5)
```

```python
# 전체컬럼에서 숫자가 들어 있는 칼럼만 추출하기, 정규표현식

df.filter(regex='\d',axis=1) # \d는 숫자를 말함
```

```python
# 컬럼 순서가 바뀐 새로운 wine_df2 만들기

group_1=['ph_5', 'sulphates', 'alcohol', 'quality_6']

group_2=['chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density']

group_3=['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar']

new_cols= group_1+ group_2 + group_3
```

```python
df2=df[new_cols]
df2
```

# 중복데이터 제거

```python
df2['density']
```

```python
df2 = df2.drop_duplicates(subset='density') # 중복된 것 제거
df2.shape
```
