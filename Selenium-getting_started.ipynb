1. 유튜브 데이터 수집하기
    - 유튜브 사이트 접속
    - '금융' 검색어 입력
    - 검색 버튼 누르기
    - 페이지다운 5번(숫자는 변경가능)누르기(pagedown)
    - 브라우저에 있는 화면(html) 저장하기
    - 내용에서 데이터 추출하기

```python
url = "https://www.youtube.com/"

browser = webdriver.Chrome("chromedriver.exe")

browser.implicitly_wait(3)

browser.get(url)

element = browser.find_element_by_id("search").send_keys('금융')

element = browser.find_element_by_xpath("//*[@id='search-icon-legacy']").click()

# 페이지 다운 실행

element=browser.find_element_by_tag_name("body")
element.send_keys(Keys.PAGE_DOWN) # page down 버튼 누르기

browser.implicitly_wait(10000)
```

```python
# 화면의 html source 받아서, BeautifulSoup 적용하기 
from bs4 import BeautifulSoup
html_source=browser.page_source

soup=BeautifulSoup(html_source, "html.parser")

soup
```
