```python
# 지역코드 파일 다운받은 주소
# "https://www.mois.go.kr/frt/bbs/type001/commonSelectBoardArticle.do?bbsId=BBSMSTR_000000000052&nttId=61552"

# 지역 코드 데이터프레임
code = pd.read_excel("KIKcd_B.20180122.xlsx")
code = code[(code["시도명"]=="서울특별시")]    
code = code[code["읍면동명"].isnull()==True]
code = code[code["시군구명"].isnull() == False]
code["gu_code"] = code["법정동코드"].astype(str).str[0:5]
code = code[['시도명', '시군구명', 'gu_code']]
code.columns = ['si','gu','gu_code']
code = code.reset_index(drop = True)

# 실거래가 오픈 API 크롤링
# 크롤링 함수만들기
def get_data(ym, lawd_cd):
    
    API_key = 'L0KZ2ICoL5OWOUXZbOCKEePlYdIOM%2F5hh5a7G3Aasau1slDVmsKI9XOoQyFsVcQrkh6ipq0%2Fm38n3PaWC2asfQ%3D%3D'
    endpoint = 'http://openapi.molit.go.kr/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcAptTradeDev'
    url = endpoint + "?serviceKey=" + API_key + "&pageNo=1" + "&numOfRows=5000" + \
            "&LAWD_CD=" + lawd_cd + "&DEAL_YMD=" + ym
    
    resultXML = urlopen(url)
    result = resultXML.read()
    xmlsoup = BeautifulSoup(result, 'lxml-xml')
    te = xmlsoup.find_all('item')

    df = pd.DataFrame()
    for t in te:
        dong = t.find('법정동').text
        apt = t.find('아파트').text
        year = t.find('년').text
        month = t.find('월').text
        day = t.find('일').text
        area = t.find('전용면적').text
        price = int(t.find('거래금액').text.replace(',',''))
        if int(month) < 10: month = '0' + month
        if int(day) < 10: day = '0' + day
        date = year + month + day
        if area <= '59': size = '중소형' # 여기 수정 필요 11! < 80 true로 인식. int()는 .때문에 오류
        elif area <= '84': size = '중대형' 
        else: size = '초대형'
            
        temp = pd.DataFrame(([[dong, apt, area, date, price, size]]),
                        columns=["dong","apt","area","date","price","size"])
        df = pd.concat([df,temp])
        
    df = df.reset_index(drop = True)
    
    return df


# 지난달, 이번달
today = date.today().strftime("%Y%m%d")[:6]
month_ago = (date.today().replace(day = 1) - timedelta(days = 1)).strftime('%Y%m%d')[:6]
time = [month_ago, today]

# 데이터 저장
df_trade = pd.DataFrame()
for t in time:
    for co in code['gu_code']:
        temp = get_data(t, co)
        df_trade = pd.concat([df_trade, temp])
df_trade = df_trade.reset_index(drop = True)
df_trade['size'] = df_trade['area'].apply(lambda x : '중소형' if int(x[0:x.find('.')]) <=59
                                          else '중대형' if int(x[0:x.find('.')]) <=84
                                          else '대형')
df_trade['pyeong'] = df_trade['area'].apply(lambda x : int(x[0:x.find('.')]) / 3.3)
df_trade['price_per_3.3'] = (df_trade['price'] / df_trade['pyeong']).round()
```
