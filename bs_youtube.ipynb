```python
import requests
from bs4 import BeautifulSoup

url= "https://www.youtube.com/results?search_query=%EA%B8%88%EC%9C%B5it"

res=requests.get(url)

soup=BeautifulSoup(res.text,"html.parser")

pp=[]

h= soup.find_all("h3",{"class": "yt-lockup-title"})

for i in h:
    b=i.text
    pp.append(b)
print(pp)
```

```python
s = soup.find_all("a",{"class": "yt-uix-sessionlink spf-link "})
po=[]

for i in s:
    s1=i.text
    po.append(s1)
print(po)
```

```python
u = soup.find_all("ul",{"class": "yt-lockup-meta-info"})
pq=[]
for i in u:
    u1=i.text
    pq.append(u1)
    
print(pq)
```

```python
p = soup.find_all("div",{"class": "yt-lockup-description yt-ui-ellipsis yt-ui-ellipsis-2"})
qq=[]
for i in p:
    p1=i.text
    qq.append(p1)
    
print(qq)
```

```python
import pandas
df=pandas.DataFrame((po, pp,pq,qq)).T

df
```
