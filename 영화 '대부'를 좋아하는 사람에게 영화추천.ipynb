# 데이터 전처리 끝

장르 Content-based filtering(CBF) 추천시스템
장르를 픽처 벡터화한 후에 행렬을 만들고, 행렬 데이터 값을 코사인 유사도로 계산

1. 장르 피처 벡터화 : genres 칼럼을 count 기반으로 피처 벡터화 변환
2. 코사인 유사도 계산 : 장르 피처 벡터화한 행렬 데이터를 코사인 유사도로 계산 -> 영화 간 유사도 정도를 숫자로 표현
3. (옵션) 새로운 수식 만들기, 평점, 관객의 VOTE수 

장르 키워드 전처리 완료
-> 영화마다 장르가 몇개씩 있겠죠

영화간 (장르) 유사도를 계산합니다.

'God father'라는 영화랑 비슷한 영화들이 나옵니다.
영화제목을 입력할 때마다 유사도 기준으로 유사한 영화를 리스트업 해줍니다. 

```python
from sklearn.feature_extraction.text import CountVectorizer

# 말뭉치
corpus = [ ' This is the first document', 'This is the second document', 'And the third one.', 'Is this the first document?', 
         'The last document?']

vect = CountVectorizer()
vect.fit(corpus)
vect.vocabulary_
```

```python
movies_df['genres_literal']=movies_df['genres'].apply(lambda x: (' ').join(x))
movies_df['genres_literal'] # 데이터를 학습하기 좋은 상황으로 만들어 줌
```

## 모델 학습

```python
count_vect=CountVectorizer(min_df=0, ngram_range=(1,2)) 

# ngram이란 단어를 하나만 학습하는 것이 아닌 앞뒤로 학습하겠다
# ngram_range : 1 <=n <=2
```

```python
genre_mat = count_vect.fit_transform(movies_df['genres_literal'])
print(genre_mat.shape)
```

# 코사인 유사도(cosine similarity) 이용해서 영화별 유사도 계산

```python
# 코사인 유사도에 의해서 4803개 영화 각각 유사한 영화들이 계산됨

from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 코사인 유사도에 의해서 4803개 영화 각각 유사한 영화들이 계산됨

genre_sim=cosine_similarity(genre_mat, genre_mat)

print(genre_sim.shape)

genre_sim
```

```python
# argsort 함수 : 정렬x, 순서 위치만 알려줌

# sort함수: 정렬을 해줌

# 유사도가 높은 영화를 순서대로 보여줌

# 0번째 영화의 경우는 유사도 순서 : 0번, 3494번, 813번, ....2401번이 유사도 최하

genre_sim_sorted_ind = genre_sim.argsort()[:, ::-1] # 전체를 -1칸 간격으로 
print(genre_sim_sorted_ind[:1])
```

# 추천함수 ver1 : 장르 코사인 유사도에 의해 영화를 추천해주는 함수

```python
def find_sim_movie_ver1(df, sorted_ind, title_name, top_n=10) :
    title_movie = df[df['title'] == title_name]
    
    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, :(top_n)]
    
    print(similar_indexes)
    
    # 2차원 데이터를 1차원으로 변환
    
    similar_indexes = similar_indexes.reshape(-1)
    
    return df.iloc[similar_indexes]
```

```python
similar_movies=find_sim_movie_ver1(movies_df, genre_sim_sorted_ind, 'The Godfather', 10)
similar_movies[['title','vote_average','genres','vote_count']]
```

## 추천 ver2를 만들건데, 평점, 평가횟수, 가중치를 적용한 공식을 만들자

```python
# 가중평점(Weighted Rating)

# (v/(v+m)R + (m/(v+m))C
 
# v : 영화별 평점 투표한 횟수(vote_count), 참여를 많이 하면 가점
 
# m : 평점을 부여하기 위한 최소 투표 횟수 -> 여기서는 상위 60%
 
# R : 개별 영화에 대한 평균 평점(vote_average)
 
# C : 전체 영화에 대한 평균 평점
```

## -> C, m 고정값. v, m 이 영화의 추천 중요도를 계산하는데 중요


```python
# 상위 6개에 해당하는 vote_count를 최소 투표 횟수인 m에 저장

C = movies_df['vote_average'].mean()
m = movies_df['vote_count'].quantile(0.6)
```

```python
print('C:', round(C, 3), 'm:', round(m,3))
```

```python
def weighted_vote_average(record):
    v = record['vote_count']
    R = record['vote_average']
    
    return ((v/(v+m)*R) + ((m/(m+v))*C))
```

```python
movies_df['weighted_vote'] = movies_df.apply(weighted_vote_average, axis=1)
```

```python
movies_df[['weighted_vote', 'title', 'vote_average', 'vote_count', 'genres']]
```

```python
def find_sim_movie_ver2(df, sorted_ind, title_name, top_n=10):
    
    title_movie = df[df['title'] == title_name]
    title_index = title_movie.index.values
    
    # top_n의 2배(10개를 추천받고 싶으면 20개를 뽑기)
    
    
    similar_indexes = sorted_ind[title_index, :(top_n * 2)]
    similar_indexes = similar_indexes.reshape(-1)
    
    # 기준 영화 index는 제외
    
    similar_indexes = similar_indexes[similar_indexes != title_index]
    
    # top_n의 2배에 해당하는 추천후보군 중에서 weighted_vote 높은 순으로 top_n개 추천
    
    return df.iloc[similar_indexes].sort_values('weighted_vote', ascending=False)[:top_n]
```

```python
find_sim_movie_ver2(movies_df, genre_sim_sorted_ind, 'The Godfather',10)
similar_movies[['title', 'vote_average', 'weighted_vote', 'genres', 'vote_count']]
```
