```python
from scipy.stats import skew

list_i = []

# 왜도

    skewness = (abs(skew(df[i], nan_policy='omit')))
    print('\n', i, '왜도 :', skewness)
    if skewness > 2:
        target = df[df[i].isna() == False][i]
        min_df = min(target)
        df[i] = np.log(target - min_df + 1)
        

    if abs(skew(df[i], nan_policy='omit')) - skewness > 0.5:
      list_i.append(i)
print(list_i)
```

```python
count = 1
plt.figure(figsize=(20,5))
for i in list_i:
  plt.subplot(1, len(list_i), count)
  plt.plot(df[i])
  plt.title(str(i))
  count += 1
plt.show()
```
