```python
url ="http://www.saramin.co.kr/zf_user/help/live"

result = requests.get(url)

soup = BeautifulSoup(result.content,"html.parser")

cat= soup.find("div",{"class": "category_list"})

cell=cat.find("table", {"class":"tbl_category_list"})

body=cell.find("tbody")

tr=body.find_all("tr")

po=[]

for i in tr:

    td=i.find("td", {"class":"category notice"})
    
    
    if td==None:
        
        td=i.find("td", {"class":"category"})
     
    
    po= po+ list(td)
    
print(po)
```
